```python
import cv2 #OpenCV
import os #Help us work with file paths
import time #To take a break between collecting images
import uuid #To help with naming image files
```

# 2. Define Images to Collect

```python
labels = ['hello', 'agree', 'love', 'thankyou', 'help']
number_imgs = 15
```

# 3. Setup Folders 

```python
IMAGES_PATH = os.path.join('Tensorflow', 'workspace', 'images', 'collectedimages')
```

```python
if not os.path.exists(IMAGES_PATH):
    if os.name == 'posix':
        !mkdir -p {IMAGES_PATH}
    if os.name == 'nt':
         !mkdir {IMAGES_PATH}
for label in labels:
    path = os.path.join(IMAGES_PATH, label)
    if not os.path.exists(path):
        !mkdir {path}
```

# 4. Capture Images

```python
for label in labels:
    cap = cv2.VideoCapture(0) # CONNECTS TO OUR WEBCAM OR CAPTURE DEVICE
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    for imgnum in range(number_imgs):
        print('Collecting image {}'.format(imgnum))
        ret, frame = cap.read()
        imgname = os.path.join(IMAGES_PATH,label,label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        cv2.imshow('frame', frame)
        time.sleep(2)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
from PIL import Image
import os

def resize_image(input_path, output_dir, size):
    # Open the image using Pillow
    with Image.open(input_path) as img:
        # Resize the image
        resized_img = img.resize(size)
        
        # Get the output path for the resized image
        output_path = os.path.join(output_dir, os.path.basename(input_path))
        
        # Save the resized image
        resized_img.save(output_path)
        
    # Return a tuple indicating success and the path of the resized image
    return True, output_path


print("Starting image resizing...")
input_dir = "C:/Users/Erik/Desktop/yeet/uni/third year/CN6000/hand recognition for sign language code/image copy resized/love"
output_dir = "C:/Users/Erik/Desktop/yeet/uni/third year/CN6000/hand recognition for sign language code/image copy resized"
size = (640, 480)

# List all JPEG files in input directory
files = [f for f in os.listdir(input_dir) if f.lower().endswith('.jpg')]
print("Found", len(files), "JPEG files in input directory")

# Loop through all files and resize them
for file in files:
    file_path = os.path.join(input_dir, file)
    success, resized_path = resize_image(file_path, output_dir, size)
    if success:
        print("Resized image saved to:", resized_path)
    else:
        print("Error resizing image:", file_path)
```
